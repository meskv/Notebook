```python
class Stack:

    def __init__(self, size):
        self.stack = [None] * size
        self.capacity = size
        self.top = -1

    def size(self):
        return self.top + 1

    def isEmpty(self):
        return self.size() == 0

    def isFull(self):
        return self.size() == self.capacity

    def peek(self):
        if self.isEmpty():
            exit(-1)
        return self.stack[self.top]


    # Operations on Stack
    def push(self, value):
        if self.isFull():
            print('Stack Overflow!! Calling exit()…')
            exit(-1)

        print('Inserting {value} into the stack…')
        self.top = self.top + 1
        self.stack[self.top] = value

    def pop(self):
        if self.isEmpty():
            print('Stack Underflow!! Calling exit()…')
            exit(-1)

        print(f'Removing {self.peek()} from the stack')

        # decrease stack size by 1 and (optionally) return the popped element
        top = self.stack[self.top]
        self.top = self.top - 1
        return top

    def traverse(self):
        for i in range(self.size()-1, -1, -1):
            print(self.stack[i], end=" ")
        print("\n")



if __name__ == '__main__':

    stack = Stack(3)

    stack.push(1)
    stack.push(2)

    stack.pop()
    stack.pop()

    stack.push(3)       # Inserting 3 in the stack

    print('Top element is', stack.peek())
    print('The stack size is', stack.size())

    stack.pop()         # removing the top element (3)

    # check if the stack is empty
    if stack.isEmpty():
        print('The stack is empty')
    else:
        print('The stack is not empty')


    stack.push(1)
    stack.push(2)
    stack.traverse()
```

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self): # Constructor
        self.head = None
        
    def isEmpty(self):
        if self.head == None:
            return True
        else:
            return False
        
    def push(self, data):
        
        if self.head ==  None:
            self.head = Node(data)
            
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
        
    def pop(self):
        if self.isEmpty():
            return None
        else:
            popped_node = self.head
            self.head = self.head.next
            popped_node.next = None
            return popped_node.data
        
    def peek(self):
        if self.isEmpty():
            return None
        else:
            return self.head.data
    
    def display(self):
        temp = self.head  # iter node
        
        if self.isEmpty():
            print("Stack underflow! \n")
        
        else:
            while(temp != None):
                print(temp.data, "->", end=" ")
                temp = temp.next
            return

# main code
st = Stack()
size = int(input("enter size of Stack: "))
for i in range(size):
    data = int(input("Enter data: "))
    st.push(data)
st.display()
print("\nTop element is: ",st.peek())
```

# Prefix to postfix

```python
output = []
precedence = {'+':1, '-':1, '*':2, '/':2, '^':3}


class Stack:
    #  for memory optimization purpose
    __slots__ = ['top', 'stack']
    
    def __init__(self):
        self.top = -1
        self.stack = []
        
     
    def isEmpty(self):
        return True if self.top == -1 else False
     

    def peek(self):
        return self.stack[-1]
     

    def pop(self):
        if not self.isEmpty():
            self.top -= 1
            return self.stack.pop()
        else:
            return "Stack Underflow!"
     
    def push(self, op):
        self.top += 1
        self.stack.append(op)
 
    def isOperand(self, char):
        return char.isalpha()

    def notGreater(self, i):
        try:
            a = precedence[i]
            b = precedence[self.peek()]
            return True if a  <= b else False
        except KeyError:
            return False
             
    def infixToPostfix(self, expression):

        for i in expression:
            if self.isOperand(i):
                output.append(i)
            
            elif i  == '(':
                self.push(i)
 
            elif i == ')':
                while( (not self.isEmpty()) and
                                self.peek() != '('):
                    a = self.pop()
                    output.append(a)
                if (not self.isEmpty() and self.peek() != '('):
                    return -1
                else:
                    self.pop()
 
            else:
                while(not self.isEmpty() and self.notGreater(i)):
                    output.append(self.pop())
                self.push(i)
 
        while not self.isEmpty():
            output.append(self.pop())
 
        print("Postfix expression: ",' '.join(output))


 
# Main Code
expression = input("Enter Infix operation: ")
stack = Stack()
stack.infixToPostfix(expression)
```

# Linked Lists


```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, value):
        new_node = Node(value)

        if(self.head):
            temp = self.head
            while (temp.next != None):
                temp = temp.next

            temp.next = new_node
        else:
            self.head = new_node

    def insertEnd(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while (temp.next != None):
            temp = temp.next

        temp.next = new_node

    def insertBetween(self, pos, value):
        if (pos == 0):
            print("Error \n")
            return

        new_node = Node(value)
        temp = self.head

        i = 1
        while(i < pos):
            temp = temp.next
            i += 1
        new_node.next = temp.next
        temp.next = new_node

    def printList(self):
        temp = self.head
        while (temp):
            print(temp.data, end=" -> ")
            temp = temp.next
        print()


# main code
l = LinkedList()
l.insert(19)  # insertion at beginning
l.insert(23)  # insertion at beginning
l.insert(31)  # insertion at beginning
l.insert(43)  # insertion at beginning
l.printList()
l.insertEnd(67)  # insertion at end
l.insertBetween(2, 51)  # position, data
l.printList()
```

# Implementation of Queue

#### Array implementation of queue

```python
class Queue:
    # constructor
    def __init__(self, capacity):
        self.front = self.size = 0
        self.rear = capacity - 1
        self.Q = [None]*capacity
        self.capacity = capacity
    
    def isFull(self):
        return self.size == self.capacity
    
    def isEmpty(self):
        return self.size == 0
    
    def enQueue(self, data):
        if self.isFull():
            print("Queue Full! \n")
            return
        self.rear = (self.rear + 1) % (self.capacity)
        self.Q[self.rear] = data
        self.size += 1
        print("% s enqueued to queue" % str(data))
        
    def deQueue(self):
        if self.isEmpty():
            print("Empty Queue! \n")
            
        print("Front item is", self.Q[self.front])
    
    def queueFront(self):
        if self.isEmpty():
            print("Empty Queue! \n")
        print("Front item is", self.Q[self.front])
    
    def queueRear(self):
        if self.isEmpty():
            print("Empty Queue! \n")
        print("Rear item is", self.Q[self.rear])
    
    
# Main code
que = Queue(30)
que.enQueue(10)
que.enQueue(20)
que.enQueue(30)
que.enQueue(40)
que.deQueue()
que.queueFront()
que.queueRear()
```
